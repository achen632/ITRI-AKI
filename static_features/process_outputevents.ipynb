```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
path    = 'MIMIC-IV v2.2/icu/outputevents.csv'
data    = pd.read_csv(path)

path    = 'MIMIC-IV v2.2/icu/d_items.csv'
items   = pd.read_csv(path)
```

```python
# Drop irrelevant columns
data = data[['subject_id', 'charttime', 'itemid', 'value']]

# Convert data into floats
data['charttime'] = pd.to_datetime(data['charttime'])
data['charttime'] = data['charttime'].apply(lambda x: x.timestamp())
```

```python
df = data.sort_values(by=['subject_id', 'itemid', 'charttime'], ascending=[True, True, False])
df.drop_duplicates(subset=['subject_id', 'itemid'], keep='first', inplace=True)

df.reset_index(drop=True, inplace=True)
df = df.pivot_table(index='subject_id', columns='itemid', values='value', aggfunc='first')
```

```python
# Rename df itemid to item labels
for i in df.columns:
    df.rename(columns={i: items.loc[items['itemid'] == i, 'label'].values[0]}, inplace=True)

# Add subject_id to df
df.insert(0, 'subject_id', df.index)
df.reset_index(drop=True, inplace=True)
```
